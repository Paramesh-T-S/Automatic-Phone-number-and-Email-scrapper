```python
import re
import pyperclip
```

Creating a regex for phone numbers

```python
PhoneRegex = re.compile(r'''
# Example phone numbers look like 415-555-1234, 555-1234 , (415) 555-1234 , 555-2222 ext 12345,ext. 12345, x12345
(((\d\d\d)|(\(\d\d\d\)))?  #area-code(optional)
(\s|-)           #first separator
\d\d\d            #first 3 digits
-                 #separator
\d\d\d\d           #last 4 digits
(((ext(\.)?\s)|x)      #Extension word-part (optional)
 (\d{2, 5}))?)
''', re.VERBOSE)
```

Creating a regex for email address

```python
emailRegex = re.compile(r'''
#Example some+._thing@(\d{2,5}))?.com

[a-zA-Z0-9_.+]+  #name part - created own character class
@                # @ symbol
[a-zA-Z0-9_.+]+  # domain name part

''', re.VERBOSE)
```

Copy the whole document or selected paragraphs in to the clipboard and get it from pyperclip

```python
text = pyperclip.paste()
```

```python
print(text)
```

Extracting mail id and Phone number from text

```python
extractedPhone = PhoneRegex.findall(text)
extractedEmail = emailRegex.findall(text)

print(extractedPhone)
print(extractedEmail)
```

```python
allPhoneNumbers = []
for phoneNumber in extractedPhone:
    allPhoneNumbers.append(phoneNumber[0])
print(allPhoneNumbers)
```

```python
print(extractedEmail)
```

Copying the extracted emails and phone numbers to the clipboard

```python
result = '\n'.join(allPhoneNumbers) + '\n' + '\n'.join(extractedEmail)
pyperclip.copy(result)
```
